# BioShake manual run notebook

Run the cells top to bottom.

If execution is interrupted while the shaker is running, execute the stop cell and then the cleanup cell before reconnecting.

```python
import logging
import time

from bioshake_driver import BioShake

DEFAULT_PORT = "/dev/ttyUSB0"
TARGET_SPEED_RPM = 500
SHAKE_DURATION_SECONDS = 5

logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s %(levelname)s %(name)s: %(message)s",
)
logger = logging.getLogger("bioshake-notebook")
```

```python
bioshake = BioShake(port=DEFAULT_PORT)
logger.info("Connected to shaker on %s", DEFAULT_PORT)
```

```python
logger.info("Homing shaker on %s", DEFAULT_PORT)
bioshake.home()
```

```python
logger.info(
    "Shaking at %s RPM for %s seconds",
    TARGET_SPEED_RPM,
    SHAKE_DURATION_SECONDS,
)
bioshake.shake(TARGET_SPEED_RPM, SHAKE_DURATION_SECONDS)
```

```python
logger.info("Stopping shaker and returning home")
bioshake.home()
```

```python
TARGET_TEMP_C = 60
TEMP_HOLD_SECONDS = 30

logger.info(
    "Heating to %d C and holding for %d seconds",
    TARGET_TEMP_C,
    TEMP_HOLD_SECONDS,
)
bioshake.set_temp(TARGET_TEMP_C, TEMP_HOLD_SECONDS)
```

```python
while True:
    actual_temp = bioshake.get_temp_actual()

    if actual_temp is None:
        raise RuntimeError("BioShake did not report an actual temperature")

    logger.info(
        "Temperature monitor: actual=%.1f C target=%.1f C",
        actual_temp,
        TARGET_TEMP_C,
    )
    time.sleep(0.5)
```

```python
logger.info("Closing ELM")
if not bioshake.close_clamp():
    raise RuntimeError("Failed to close the clamp")
```

```python
logger.info("Opening ELM")
if not bioshake.open_clamp():
    raise RuntimeError("Failed to open the clamp")
```

```python
bioshake.disconnect()
logger.info("Disconnected from shaker")
```
